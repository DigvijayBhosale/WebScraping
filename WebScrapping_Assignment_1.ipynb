# ICC_Circket 

```python
top_odi_10=[]
top_odi_batsman=[]
top_odi_bowling=[]
top_10_team_women=[]
top_odi_batsman_women=[]
top_odi_bowling_women=[]

URLst=['https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling',
       'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting',
       'https://www.icc-cricket.com/rankings/mens/team-rankings/odi',
      'https://www.icc-cricket.com/rankings/womens/player-rankings/odi/bowling',
       'https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting',
       'https://www.icc-cricket.com/rankings/womens/team-rankings/odi']

a=0
for url in URLst:
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    a += 1
    var='title'+str(a)
    title=soup.find_all('td') #,class_='rankings-block__banner')
    if var == 'title1':
        for i in title:
            top_odi_bowling.append(i.get_text().replace("\n",""))
    elif var == 'title2':
        for i in title:
            top_odi_batsman.append(i.get_text().replace("\n",""))
    elif var == 'title3':
        for i in title:
            top_odi_10.append(i.get_text().replace("\n",""))
    elif var == 'title4':
        for i in title:
            top_odi_bowling_women.append(i.get_text().replace("\n",""))        
    elif var == 'title5':
        for i in title:
            top_odi_batsman_women.append(i.get_text().replace("\n",""))        
    else:
        for i in title:
            top_10_team_women.append(i.get_text().replace("\n",""))
        

def create_arr(lst):
    a = 0
    c = []
    for i in range(0,55,5):
        c.append(lst[a:i])
        a = i
    c.remove([])
    for i in c:
        del i[0]
    counter = 0
    for i in c:
        counter += 1
        i.append(counter)
    return c

top_odi_batsman = create_arr(top_odi_batsman)
top_odi_bowling = create_arr(top_odi_bowling)
top_odi_10 = create_arr(top_odi_10)
top_10_team_women = create_arr(top_10_team_women)
top_odi_batsman_women = create_arr(top_odi_batsman_women)
top_odi_bowling_women = create_arr(top_odi_bowling_women)



Top10Teams=pd.DataFrame(top_odi_10,columns = ['Rank','Team','Matches','Points','Rating'])
Top10Batsman=pd.DataFrame(top_odi_batsman,columns = ['Player','Team','Points','Best','Rank'])
Top10Bowler=pd.DataFrame(top_odi_bowling,columns = ['Player','Team','Points','Best','Rank'])
Top10Teams_Women=pd.DataFrame(top_10_team_women,columns = ['Rank','Team','Matches','Points','Rating'])
Top10Batsman_Women=pd.DataFrame(top_odi_batsman_women,columns = ['Player','Team','Points','Best','Rank'])
Top10Bowler_Women=pd.DataFrame(top_odi_bowling_women,columns = ['Player','Team','Points','Best','Rank'])




    
```

# AMAZON Mobile List

```python
Mobile=[]
Mobstrrt=[]
MImages=[]
MPrice=[]

page=requests.get('https://www.amazon.in/s?k=smartphone+under+20000+rs+with+4%2B+rating&rh=p_72%3A1318477031&dc&qid=1610695524&rnid=1318475031&ref=sr_nr_p_72_2')
soup=BeautifulSoup(page.content,'html.parser')
mobilename=soup.find_all('a')
for i in mobilename:
    mobname=soup.find_all('span',class_='a-size-medium a-color-base a-text-normal')
for i in mobname:
    Mobile.append(i.get_text().replace("\n",""))
price=soup.find_all('a')    
for i in price:
    mobpr=soup.find_all('span',class_='a-price-whole')
for i in mobpr:
    MPrice.append(i.get_text().replace("\n",""))
ratings=soup.find_all('i')
for i in ratings:
    starrating=soup.find_all('span',class_='a-icon-alt')
for i in starrating[4:]:
    Mobstrrt.append(i.get_text().replace("\n",""))
img=soup.find_all('img',class_="s-image")
for i in img:
    MImages.append(i.get('src'))
```

```python
result=pd.DataFrame({'Mobile_name': Mobile,'Price':MPrice,'Rating':Mobstrrt,'Mobile_Image':MImages})
result
```

# BookName with all the details

```python
page=requests.get('https://bookpage.com/reviews')
soup=BeautifulSoup(page.content,'html.parser')
name=soup.find_all('h4',class_='italic')
Book_Name=[]
for i in name:
    Book_Name.append(i.get_text().replace("\n",""))

author=soup.find_all('p',class_='sans bold')
Author_name=[]
for i in author:
    Author_name.append(i.get_text().replace("\n",""))

genre=soup.find_all('p',class_='genre-links hidden-phone')
Genre=[]
for i in genre:
    Genre.append(i.get_text().replace("\n",""))
    
desc=soup.find_all('p',class_='excerpt')
Description=[]
for i in desc:
    Description.append(i.get_text().replace("\n",""))


bookresult=pd.DataFrame({'Book_Name': Book_Name,'Author':Author_name,'Genre':Genre,'Description':Description})
bookresult
```

# Weather Forecast

```python
pages=['https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YAOtVtAzbIV']

day=[]
temp=[]
description=[]
desc=[]
city=[]

for i in pages:
    page=requests.get(i)
    soup=BeautifulSoup(page.content,'html.parser')
    dayt=soup.find_all('div',class_='col-sm-2 forecast-label')
    for i in dayt:
        day.append(i.get_text().replace("\n",""))
    output=soup.find_all('div',class_='col-sm-10 forecast-text')
    for i in output:
        description.append(i.get_text().replace("\n",""))
    temp_op=soup.find_all('p',class_='temp')    
    for i in temp_op:
        temp.append(i.get_text().replace("\n",""))
    desc_op=soup.find_all('p',class_='short-desc')
    for i in desc_op:
        desc.append(i.get_text().replace("\n",""))
    ct=soup.find_all('h2',class_='panel-title')
    for i in ct:
        city.append(i.get_text().replace("\n",""))
        
print( '\t City Weather Forecast : ',city[0])
forecastresult=pd.DataFrame({'Day': day[0:9],'Tempture':temp,'Short Description':desc,'Description':description[0:9]})
forecastresult
    
```
