# Wikipedia Most Viewed YouTube videoes

```python
def youtube_video_list(url):
    rank = []
    Name = []        
    Artist = []        
    views = []        
    uploadedon = []
    driver=webdriver.Chrome('C://Users//735184//Downloads//chromedriver_win32//chromedriver.exe')
    driver.get(url)
    for i in range(1,15):
        rnk = ("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr["+str(i)+"]/td[1]")
        nm = ("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr["+str(i)+"]/td[2]")
        atst = ("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr["+str(i)+"]/td[3]")
        vw = ("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr["+str(i)+"]/td[4]")
        dt = ("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr["+str(i)+"]/td[5]")
        try:
            a=driver.find_elements_by_xpath(rnk)
            for i in a:
                rank.append(i.text.replace('.',''))
        except NoSuchElementException :
            rank.append("--NA--")
        try:
            b=driver.find_elements_by_xpath(nm)
            for i in b:
                Name.append(i.text.replace('.',''))
        except NoSuchElementException :
            Name.append("--NA--")
        try:
            c=driver.find_elements_by_xpath(atst)
            for i in c:
                Artist.append(i.text.replace('.',''))
        except NoSuchElementException :
            Artist.append("--NA--")
        try:
            d=driver.find_elements_by_xpath(vw)
            for i in d:
                views.append(i.text+ ' Billion')
        except NoSuchElementException :
            views.append("--NA--")
        try:
            e=driver.find_elements_by_xpath(dt)
            for i in e:
                uploadedon.append(i.text.replace('.',''))
        except NoSuchElementException :
            uploadedon.append("--NA--")

    driver.quit()

    datafrm=pd.DataFrame({"Rank":rank,"Video Name":Name,"Artist":Artist,"Views":views,"Uploaded Date":uploadedon})
    return datafrm
    
 
        
```

```python
youtube_video_list('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

# BCCI Match Details

```python
driver_bcci=webdriver.Chrome('C://Users//735184//Downloads//chromedriver_win32//chromedriver.exe')
driver_bcci.get('https://www.bcci.tv/')
title = []
match = []
location = []
datetime = []
urls=[i.get_attribute("href") for i in driver_bcci.find_elements_by_xpath("//div[@class='drop-down__options']/ul/li/a")]
for i in urls:
    if re.search('international/fixtures',i):
        driver_bcci.get(i)
```

# Naukri Data Architect

```python
driver=webdriver.Chrome('C://Users//735184//Downloads//chromedriver_win32//chromedriver.exe')
driver.get("https://www.naukri.com/")
keywords = driver.find_element_by_name("keyword")
keywords.clear()
keywords.send_keys('Data Architect')
time.sleep(5)
location =  driver.find_element_by_name("location")
search_btn = driver.find_elements_by_xpath("//button[@class='btn']")
search_btn[0].click()


joblist = []
company = []        
location = []        
experience = []        
salary = []
time.sleep(2)
job_title=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
cmp_name=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
lcn_name=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
exp_yr=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
sal=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi salary']")
tags=[job_title,cmp_name,lcn_name,exp_yr,sal]
for i in tags:
    if i == job_title:
        for t in i:
            joblist.append(t.text)
    elif i == cmp_name:
        for t in i:
            company.append(t.text)
    elif i == lcn_name:
        for t in i:
            location.append(t.text)
    elif i == exp_yr:
        for t in i:
            experience.append(t.text)
    else:
        for t in sal:
            salary.append(t.text)
time.sleep(2)
driver.quit()
Data_Scnt_Blr=pd.DataFrame({'Job Title':joblist,'Company Name':company,'Location':location,'Experience':experience,'Salary':salary})
Data_Scnt_Blr

 
```
