# Created one class for selenium example  and inside that class multiple methods are defined to solve Selenium assignments

Note: To populate first 10 records I have not used list[0:10]. Menthods are returning all the details instead of first 10 records.

```python
class selenium_exple:
    '''
    To scrape data for Keywords for multiple location.
    Parameter details need to pass this class.    
    Keywords : Example - Data Analyst, Python Developer, Data Architect etc.
    Location : Example - Pune ,Mumbai ,Bangalore etc or default is blank
    drv      : Default is Chrome driver and we can pass any webdriver's exe file
    
    '''
    def __init__(self, search_key, location='' ,drv='C://Users//735184//Downloads//chromedriver_win32//chromedriver.exe'):
        self.search_key = search_key
        self.location = location
        self.drv = drv
    def chrm_driver(self):
        driver_chm = webdriver.Chrome(self.drv)
        return driver_chm
    
    def noukri_srch(self):
        driver=self.chrm_driver()
        driver.get("https://www.naukri.com/")
        keywords = driver.find_element_by_name("keyword")
        keywords.clear()
        keywords.send_keys(self.search_key)
        location =  driver.find_element_by_name("location")
        if self.location != '':
            location.clear()
            location.send_keys(self.location)
        search_btn = driver.find_elements_by_xpath("//button[@class='btn']")
        search_btn[0].click()
        return driver
        
    def noukri_asg1(self):
        joblist = []
        company = []        
        location = []        
        experience = []        
        salary = []
        n_driver=self.noukri_srch()
        time.sleep(2)
        job_title=n_driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
        cmp_name=n_driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
        lcn_name=n_driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
        exp_yr=n_driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
        sal=n_driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi salary']")
        tags=[job_title,cmp_name,lcn_name,exp_yr,sal]
        for i in tags:
            if i == job_title:
                for t in i:
                    joblist.append(t.text)
            elif i == cmp_name:
                for t in i:
                    company.append(t.text)
            elif i == lcn_name:
                for t in i:
                    location.append(t.text)
            elif i == exp_yr:
                for t in i:
                    experience.append(t.text)
            else:
                for t in sal:
                    salary.append(t.text)
        time.sleep(2)
        n_driver.quit()
        Data_Scnt_Blr=pd.DataFrame({'Job Title':joblist,'Company Name':company,'Location':location,'Experience':experience,'Salary':salary})
        return Data_Scnt_Blr
        
    def noukri_asg2(self):
        job = []
        location = []
        company = []
        description = []
        exp = []
        asg2_driver=self.noukri_srch()
        time.sleep(2)
        links=asg2_driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
        for i in links:
            i.click()
        parenttab = asg2_driver.current_window_handle
        alltabs = asg2_driver.window_handles
        for tab in alltabs:
            if tab != parenttab:
                asg2_driver.switch_to.window(tab)
                joblst = asg2_driver.find_elements_by_xpath("//h1[@class='jd-header-title']")
                for i in joblst:
                    job.append(i.text)
                cmp = asg2_driver.find_elements_by_xpath("//div[@class='jd-header-comp-name']")
                for i in cmp:
                    company.append(i.text)
                company = [string for string in company if string != ""]
                    
                loctn = asg2_driver.find_elements_by_xpath("//span[@class='location ']")
                for i in loctn:
                    location.append(i.text)
                desc = asg2_driver.find_elements_by_xpath("//div[@class='dang-inner-html']")
                for i in desc:
                    description.append((i.text).replace("\n",""))
                experc = asg2_driver.find_elements_by_xpath("//div[@class='exp']")
                for i in experc:
                    exp.append(i.text)
        time.sleep(2)
        asg2_driver.quit()
        Data_Scnt_Blr=pd.DataFrame({'Job Title':job,'Company Name':company,'Location':location,'Experience':exp,'Description':description})
        return Data_Scnt_Blr
        
    def noukri_asg3(self):
        joblist = []
        location = []
        company = []
        salary = []
        experience = []
        adriver = self.noukri_srch()
        adriver.maximize_window()
        time.sleep(2)
        loc_fltr = adriver.find_elements_by_xpath("//span[@title='Delhi/NCR']") #"//span[@title='Delhi/NCR']")
        time.sleep(2)
        for i in loc_fltr:
            i.click()
        time.sleep(2)
        sal_fltr = adriver.find_elements_by_xpath("//span[@title='3-6 Lakhs']")
        for i in sal_fltr:
            i.click()
        time.sleep(2)
        job_title=adriver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
        cmp_name=adriver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
        lcn_name=adriver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
        exp_yr=adriver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
        sal=adriver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi salary']")
        tags=[job_title,cmp_name,lcn_name,exp_yr,sal]
        for i in tags:
            if i == job_title:
                for t in i:
                    joblist.append(t.text)
            elif i == cmp_name:
                for t in i:
                    company.append(t.text)
            elif i == lcn_name:
                for t in i:
                    location.append(t.text)
            elif i == exp_yr:
                for t in i:
                    experience.append(t.text)
            else:
                for t in sal:
                    salary.append(t.text)
        time.sleep(5)
        adriver.quit()
        Data_Scnt_Delhi_Noida=pd.DataFrame({'Job Title':joblist,'Company Name':company,'Location':location,'Experience':experience,'Salaray':salary})
        return Data_Scnt_Delhi_Noida
         

        
```

# Flipkart Data Scrapping : Main Class 

```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd



class selenium_flkrt():
    '''
    To scrape data for multiple products at flipkart.
    Parameter details need to pass this class.    
    Product  : Example - Mobiles,Sunglasses,laptop etc.
    URL      : Example - any website by default it is Flipkart
    drv      : Default is Chrome driver and we can pass any webdriver's exe file
    
    '''
    def __init__(self, search_key,url='https://www.flipkart.com/',drv='C://Users//735184//Downloads//chromedriver_win32//chromedriver.exe'):
        self.search_key = search_key
        self.url = url
        self.drv = drv
    def flkrt_driver(self):
        driver = webdriver.Chrome(self.drv)
        driver.get(self.url)
        driver.maximize_window()
        time.sleep(2)
        if driver.find_elements_by_xpath("//div[@class='_2QfC02']") is not None:
            time.sleep(2)
            driver.find_elements_by_xpath("//button[@class='_2KpZ6l _2doB4z']")[0].click()
            time.sleep(2)
        product = driver.find_elements_by_xpath("//input[@class='_3704LK']")
        product[0].send_keys(self.search_key)
        time.sleep(2)
        driver.find_elements_by_xpath("//button[@class='L0Z3Pu']")[0].click()
        return driver
    def product_flpkrt(self):
        '''
        Any Product Serach on Flipkart like Sunglasses or Sneakers ect.
        
        '''
        desc = []
        brand = []
        price = []
        discount = []
        driver_prd=self.flkrt_driver()
        time.sleep(2)
        for i in driver_prd.find_elements_by_xpath("//div[@class='_3879cV']"):
            if i.text == '10% or more':
                time.sleep(2)
                i.click()
        for i in range(3):    #Loop to click on next button 3 times
            time.sleep(2)
            for i in driver_prd.find_elements_by_xpath("//a[@class='IRpwTa' or contains(@class, 'IRpwTa _2-ICcC')]"):
                desc.append(i.text)
            for i in driver_prd.find_elements_by_xpath("//div[@class='_2WkVRV']"):
                brand.append(i.text)
            for i in driver_prd.find_elements_by_xpath("//div[@class='_30jeq3']"):
                price.append(i.text)
            for i in driver_prd.find_elements_by_xpath("//div[@class='_3Ay6Sb']"):
                discount.append(i.text) 
            for nxt in driver_prd.find_elements_by_xpath("//a[@class='_1LKTO3']"):  #Click on Next Button
                if nxt.text == 'NEXT':
                    time.sleep(2)
                    nxt.click()
        driver_prd.quit()
        product_flipkart=pd.DataFrame({'Brand':brand,'Description':desc,'Price':price,'Discount':discount})
        return product_flipkart
    def iphone_rvflkrt(self):
        starrt = []
        rvw = []
        desc = []
        driver = webdriver.Chrome(self.drv)
        driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb-includesearpods-poweradapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKC%20TSVZAXUHGREPBFGI&marketplace")
        driver.maximize_window()
        time.sleep(2)
        driver.find_elements_by_xpath("//div[@class='_3UAT2v _16PBlm']")[0].click()
        time.sleep(2)
        for i in range(10):
            time.sleep(3)
            for i in driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq' or contains(@class, '_3LWZlK _1rdVr6 _1BLPMq')]"):
                starrt.append(i.text)
            for i in driver.find_elements_by_xpath("//p[@class='_2-N8zT']"):
                rvw.append(i.text)
            for i in driver.find_elements_by_xpath("//div[@class='t-ZTKy']"):
                desc.append((i.text).replace("\n",""))
            time.sleep(2)
            for i in driver.find_elements_by_xpath("//a[@class='_1LKTO3']"):
                if i.text == 'NEXT':
                    time.sleep(2)
                    i.click()
        time.sleep(2)
        driver.quit()
        iphone_review=pd.DataFrame({'Star Ratings':starrt,'Review':rvw,'Description':desc})
        return iphone_review
        
        
    
```
