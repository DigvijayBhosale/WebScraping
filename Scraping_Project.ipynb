```python
from tkinter import *
import tkinter as tk
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup
import requests


```

```python
window = tk.Tk()
frame=Frame(window, width=700, height=900)
frame.pack()
window.title("Web Scrapping Project")
#window.configure(bg="white")


def wikipedia():
    page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
    soup=BeautifulSoup(page.content,'html.parser')
    title=soup.find_all(re.compile('^h[1-6]$'))
    g=f"-----WikiPedia header-----\n"
    ans.configure(text=g)
    a=""
    for i in title:
        r=i.get_text().replace("\n","")
        a=a+'\n'+r
           
    ans2=tk.Label(window,text=a,fg="black")
    ans2.configure(text=a)
    ans2.place(x=70,y=90)
    
def IMDB_TOP_25():
    movies=[]
    page=requests.get('https://www.imdb.com/list/ls000060680/')
    soup=BeautifulSoup(page.content,'html.parser')
    title=soup.find_all('h3',class_='lister-item-header')
    g=f"-----IMBD Top 25 Movies as per Ratings -----\n"
    ans.configure(text=g)
    a=""
    for i in title[0:25]:
        r=i.get_text().replace("\n","")
        a=a+'\n'+r
          
    ans2=tk.Label(window,text=a,fg="black")
    ans2.configure(text=a)
    ans2.place(x=70,y=90)

def IMDB_BOLY_TOP_25():
    movies=[]
    page=requests.get('https://www.imdb.com/list/ls009997493/')
    soup=BeautifulSoup(page.content,'html.parser')
    title=soup.find_all('h3',class_='lister-item-header')
    g=f"-----IMBD Top 25 Bollywood Movies as per Ratings -----\n"
    ans.configure(text=g)
    a=""
    for i in title[0:25]:
        r=i.get_text().replace("\n","")
        a=a+'\n'+r

    ans2=tk.Label(window,text=a,fg="black")
    ans2.configure(text=a)
    ans2.place(x=70,y=90)    
    

    

l = Label(window, text="Web Scrapping Projects ")
l.place(x=155, y=5)
#l.pack()


        
button_wikipedia = Button(window, text="Question 1: wikipedia", command=wikipedia)
button_wikipedia.place(x=5, y=30)
ans=tk.Label(window,text="",bg="black",fg="yellow")
ans.place(x=50,y=70)

button_IMDB_BOLY_TOP_25 = Button(window, text="Question 2: IMDB_Bolly_25", command=IMDB_BOLY_TOP_25)
button_IMDB_BOLY_TOP_25.place(x=155, y=30)
ans=tk.Label(window,text="",bg="black",fg="yellow")
ans.place(x=50,y=70)

button_IMDB_TOP_25 = Button(window, text="Question 3: IMDB_TOP_25", command=IMDB_TOP_25)
button_IMDB_TOP_25.place(x=355, y=30)
ans=tk.Label(window,text="",bg="black",fg="yellow")
ans.place(x=50,y=70)





window.mainloop()
```
