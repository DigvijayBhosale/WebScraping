```python
from bs4 import BeautifulSoup
import requests
import warnings
warnings.filterwarnings('ignore')
import re
```

```python
pages=['https://www.amazon.in/s?k=smartphone+under+20000+rs+with+3%2B+rating&rh=p_72%3A1318477031&dc&qid=1610695524&rnid=1318475031&ref=sr_nr_p_72_2']

Mobile=[]
Mobstrrt=[]
MImages=[]
MPrice=[]

for i in pages:
    page=requests.get(i)
    soup=BeautifulSoup(page.content,'html.parser')
    mobilename=soup.find_all('a')
    for i in mobilename:
        mobname=soup.find_all('span',class_='a-size-medium a-color-base a-text-normal')
    for i in mobname:
        Mobile.append(i.get_text().replace("\n",""))
    price=soup.find_all('a')    
    for i in price:
        mobpr=soup.find_all('span',class_='a-price-whole')
    for i in mobpr:
        MPrice.append(i.get_text().replace("\n",""))
    ratings=soup.find_all('i')
    for i in ratings:
        starrating=soup.find_all('span',class_='a-icon-alt')
    for i in starrating[4:]:
        Mobstrrt.append(i.get_text().replace("\n",""))
    img=soup.find_all('img',class_="s-image")
    for i in img:
        MImages.append(i.get('src'))


print(len(Mobile))
print(len(MPrice))
print(len(Mobstrrt))
print(len(MImages))
```

```python
import pandas as pd
result=pd.DataFrame({'Mobile_name': Mobile,'Price':MPrice,'Rating':Mobstrrt,'Mobile_Image':MImages})
result
```
